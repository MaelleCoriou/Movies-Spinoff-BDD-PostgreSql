# PROJET FLOUPICS - CREATION BDD PostGreSQL

Vous avez fait le bon choix en choisissant PostGreSql. 

Il s'agit d'un véritable moteur de base de données, SQL est un langage de programmation avec lequel nous travaillons pour communiquer avec une base de données. 
Il s'agit d'une base de données relationnelle objet, ce qui signifie que la relation entre les deux tables est conservée. Toute base de données a un format de table qui est intégré aux lignes et aux colonnes.

Ses avantages :

   - Il gère les données dans une base de données relationnelle.
   - Il est très robuste et puissant.
   - Open source.
   - Beaucoup de soutien communautaire.

---

## CREATION BDD PostGreSQL

Après avoir installé PostGreSql :
     
  - Lancez l'application PgAdmin : une fenêtre web va s'ouvrir
  - Votre mot de passe vous sera demandé : vous aurez axxès aux serveurs   

Pour créer une nouvelle base de données :
     
   - Clique Droit sur Database dans l'arboressence PostgreSql > create > Database
   - Dans la nouvelle fenêtre > Create Database > Data Base 
     ---> Saisir le nom de votre Base de Données 

**La base de données est créée !**
